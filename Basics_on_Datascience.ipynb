```python
import pandas as pd
import seaborn as sbn
```

```python
df=pd.read_excel('/content/Sample - Superstore.xls')
```

```python
df.head(5)
```

To view the lastt 5 elements

```python
df.tail(5)
```

Grouping the columns by state and finding the sum of each state

```python
df.groupby('State')['Profit'].sum()
```

```python
df.groupby('State')['Sales'].sum()
```

**find out the top 3 *states* with maximum  loss**

```python
df.groupby('State')['Sales'].sum().sort_values(ascending=True).head(3)
```

**Least profit making Category**

```python
df.groupby('Category')['Profit'].sum().sort_values(ascending=True).head(1)
```

**count the number of unique orders for different categories**

```python
df.groupby('Category')['Order ID'].nunique().reset_index()
```

```python
df.corr()
```

```python
x = df[['Sales','Quantity',	'Discount',	'Profit']].corr()
```

```python
x
```

**Now lets visually represent this correlation using SEABORN library**

```python
sbn.heatmap(x,annot=True)
```

```python
category_order= df.groupby(['Category', 'Sub-Category'])['Order ID'].nunique().reset_index()
```

```python
category_order
```

```python
sbn.catplot(
    data=category_order, kind="bar",
    x="Category", y="Order ID", hue='Sub-Category')
```
